```python
# Copy these files
# Copy the data files
# And the HyPhy results


# Data directory E:\BUSTEDS-MH-UpdateSnakefile\data\Enard

DataDirectory = os.path.join("..", "data", "Enard")
ResultsDirectory = os.path.join("..", "analysis", "Enard")
OutDirectory  = os.path.join("..", "interrogate")
path = os.path.join(OutDirectory)

for n, file in enumerate(df_minus_plus["File"].to_list()):
    print(n+1, file)
    get_alignment = os.path.join(DataDirectory, file + ".fa")
    # dir interrogate
    src = get_alignment
    dst = os.path.join(OutDirectory, file + ".fa")
    shutil.copyfile(src, dst)
    
    for analysis_json in [".BUSTED.json", ".BUSTED.MH.json", ".BUSTED.MH.noS.json", ".BUSTED.S.json"]:
        src = os.path.join(ResultsDirectory, file + analysis_json)
        dst = os.path.join(OutDirectory, file + analysis_json)
        shutil.copyfile(src, dst)
    #end for
#end for
        
```
