## Helper functions

```python
def process_viz(df, label, pval_Threshold=0.05, parameters=["Simulated_DH", "Simulated_TH",  'Simulated_THSI']):
    df_dict, count = {}, 0
    df_export = []
    for model in ["BUSTEDS-MH", "BUSTEDS"]:
        print("# Examining model:", model)
        df2 = df[df["Method"] == model]
        for parameter in parameters:
            #lst_holder = parameters
            print("# Parameter:", parameter)
            #parameter_range = set(df[parameter])
            parameter_range = [0.05, 0.25, 0.1, 0.5]
            for value in parameter_range:
                print("# Parameter value:", value)
                df_holder = df2[df2[parameter] == value] # dataframe of this single DH rate
                
                if float(value) == 0.05:
                    for param in parameters:
                        if param == parameter: 
                            continue
                        #end if
                        print("# Filtering:", param)
                        df_holder = df_holder[df_holder[param] == 0.0]
                        print(len(df_export), df_holder.shape[0])
                    #end for
                    #return df_holder
                #end if
                
                df_export.append(df_holder)
                n =  df_holder.shape[0]  # Number of items
                b =  df_holder[df_holder["LRT p-value"] <= pval_Threshold].shape[0] # number of items below threshold
                fraction = (b / n) # calculate simple fraction        
                df_dict[count] = {label: round(fraction, 3), 
                                  parameter: str(round(value,2)), 
                                  "Method": model, 
                                  "Parameter": parameter, 
                                  "Parameter_Value": str(round(value,2)),
                                  "Num.Items.In.Bin": str(n)}
                count += 1
            #end for
            #count += 1
        #end for
    #end outer for
    # Process to df
    
    df_df_dict = pd.DataFrame.from_dict(df_dict, orient="index")
    df_df_dict.index += 1
    return df_df_dict, df_export
#end method
```

## Data process

```python
df_plot_w1, df_export = process_viz(df2[df2["Simulated_Omega3"] == 1.0], "FPR")
#df_plot_w2_077 = process_viz(df[df["Simulated_Omega3"] == 2.077], "Power")
#df_plot_w6 = process_viz(df[df["Simulated_Omega3"] == 6.0], "Power")
```
